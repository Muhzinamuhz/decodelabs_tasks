***Loading the Dataset***

```python
import pandas as pd

df = pd.read_excel("Dataset for Data Analytics (1).xlsx")

df.head()
```



The dataset is loaded using Pandas. The head() function displays the first 5 rows of the dataset to understand its structure and contents.

***Basic Statistics***

```python
df.describe()
```

This provides summary statistics like mean, count, minimum, and maximum values.

***Checking Missing Values***

```python
df.isnull().sum()
```

This code checks for missing (null) values in each column of the dataset. It helps identify incomplete data that may affect analysis.

***Product Analysis***

```python
df['Product'].value_counts()
```

This shows the number of times each product appears in the dataset. It helps identify the most frequently sold products.

***Payment Method Analysis***

```python
df['PaymentMethod'].value_counts()
```

This shows the number of orders in each status (e.g., shipped, delivered, cancelled), helping analyze order fulfillment performance.

***Order Status Analysis***

```python
df['OrderStatus'].value_counts()
```

This code counts the number of orders in each category such as Shipped, Delivered, Cancelled, and Returned. It helps understand how orders are processed and completed.

***Total Sales Calculation***

```python
df['TotalPrice'].sum()
```

This calculates the total revenue generated from all orders in the dataset.

***Order Status Distribution Visualization***

```python
import matplotlib.pyplot as plt
df['OrderStatus'].value_counts().plot(kind='bar')
plt.show()
```

This bar chart visually represents the distribution of order statuses, making it easier to compare how many orders are delivered, shipped, cancelled, or returned.

***Product Distribution Visualization***

```python
import matplotlib.pyplot as plt

df['Product'].value_counts().plot(kind='bar')
plt.show()
```

This bar chart visualizes the distribution of products sold, making it easier to compare product popularity.

***Outlier Detection using Boxplot***

```python
import matplotlib.pyplot as plt

df.boxplot(figsize=(10,6))
plt.show()
```

This boxplot helps identify outliers in numerical columns. Values that lie far outside the normal range are considered outliers.
